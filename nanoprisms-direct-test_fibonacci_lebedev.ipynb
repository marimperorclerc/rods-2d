```python
%matplotlib widget
from nanoprisms_fibonacci import *
import numpy as np

from matplotlib import pyplot as plt
```

```python
sld=1
sld_solvent=0
nsides=5
Rave=600
L=3500
norder=31 # order lebedev
npoints = 300 # fibonacci
```

```python
plot_fibonacci_sphere(npoints)
```

```python
from matplotlib import pyplot as plt

qs = np.logspace(-4,-1,300,base=10)

plt.figure()
plt.loglog(qs,Iq_lebedev(qs,sld,sld_solvent,nsides,Rave,L,norder)/volume(nsides,Rave,L)**2 ,label='Lebedev' )
plt.loglog(qs,Iq_fibonacci(qs,sld,sld_solvent,nsides,Rave,L,npoints)/volume(nsides,Rave,L)**2, label='Fibonacci' )

plt.legend()
```

Les mthodes Lebedev (ordre=131) et FIbonacci (npoints= 300) donnent de rsultats quivalents

```python
plt.figure()
npoints=[100,250,500,750,1000]
for npoints in npoints:
    plt.loglog(qs,Iq_fibonacci(qs,sld,sld_solvent,nsides,Rave,L,npoints)/volume(nsides,Rave,L)**2, label=f'{npoints} points' )
plt.legend()
```

La mthode de FIbonacci semble donner une solution stationnaire pour nobmre de points suprieur  500. 

Si on considre un nombre de points infrieur  500, la courbe SAXS semble dvier aux grandes valeurs de q. 
D'autre part, le fait que la quadrature de Lebedev d'ordre 131 donne une solution quivalente  celle obtenue avec une grille Fibonacci de 300 points, on pourrait penser que la mthode de Fibonacci est plus prcise,  condition de considrer un nombre de points sur la sphre unit suprieur  500.
