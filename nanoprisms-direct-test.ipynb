```python
from nanoprisms import *
import numpy as np

from matplotlib import pyplot as plt
```

```python
sld=1
sld_solvent=0
nsides=5
Rave=600
L=3500
norder=31
```

```python
from matplotlib import pyplot as plt

qs = np.logspace(-4,-1,200,base=10)


plt.plot(qs,[Iq(qi,sld,sld_solvent,nsides,Rave,L,norder)/volume(nsides,Rave,L)**2 for qi in qs])
plt.loglog()
plt.show()
```
